### Dictionary methods
#### keys() - It is a method in dictionary which retuens all of the keys present inside our dictionary 
#### values()  - It will return all values inside our dictionary
#### items() - It will be used to return key value pairs together
#### get()  -  It will be used for returning the values for the key we are going to pass inside our get method 

```python
dictionary = {"name": "Harini", "fruit": "orange", "vegetable": "potato","List_items" :[1,2,3,4,5], "Number": 10, "Boolean": True,"Float": 3.14}
print(dictionary)
```

```python
 print(dictionary.keys()) #Returns a view object containing the keys of the dictionary
```

```python
print(dictionary.values()) #Returns a view object containing the values of the dictionary
```

```python
print(dictionary.items()) #Returns a view object containing the key-value pairs of the dictionary
```

```python
print(dictionary.get("name")) #Returns the value associated with the key "name" using the get() method
```

#### Shallow copy - It is used for creating a copy of the original dictionary in a way such that the changes we are going to make with the original dictionary won't be getting reflected into our copied dictionary.
#### Shallow copy is being done by copy()

```python
dict_1 = {"name": "Harini", "age": 23, "city": "Vizag", "qualification": "B.Tech"}
dict_2 = dict_1
print(dict_2)
```

```python
dict_1["name"] = "Hari"
print(dict_1) #Updating the value associated with the key "name" in dict_1
print(dict_2) #Since dict_2 is referencing the same dictionary as dict_1, the change is reflected in dict_2 as well
```

```python
dict_3 = dict_1.copy() #Creating a shallow copy of dict_1 and assigning it to dict_3
print(dict_3) #Printing dict_3 to verify that it contains the same key-value pairs as dict_1
```

```python
dict_1["name"] = "Harry"
print(dict_1) #Updating the value associated with the key "name" in dict_1
print(dict_3) #Since dict_3 is a copy of dict_1, the change in dict_1 does not affect dict_3
```

### Nested Dictionary
#### This will be used for keeping a dictionary inside another dictionary.

```python
# I want to store the information for multiple students like name, marks, and grade. I can use a list of dictionaries to achieve this. Each dictionary will represent a student and contain their information.
students = {
    "Student_1": {"name": "Alice", "marks": 85, "grade": "A"},
    "Student_2": {"name": "Bob", "marks": 75, "grade": "B"},
    "Student_3": {"name": "Charlie", "marks": 90, "grade": "A"}
}
```

```python
print(students)
```

```python
# Iterate over the students dictionary(nested dictionary)
for outer_key, outer_value in students.items():
    print(f"Outer Key: {outer_key} and Outer Value: {outer_value}")
    for key_1,value_1 in outer_value.items():
        print(f"Inner Key: {key_1} and Inner Value: {value_1}")
```

#### Dictionary comprehension
##### Dictionary comprehension helps us with using looping statement along with conditional statement inside our dictionary.
##### Syntax for dictionary comprehension : {expresiion for item in iterable if condition}

```python
Comprehensive = {x**2 for x in range(1, 11)} #Creating a set of squares of numbers from 1 to 10 using set comprehension
print(Comprehensive)
```

```python
Comprehensive_dict = {x : x**2 for x in range(1, 11)} #Creating a dictionary of squares of numbers from 1 to 10 using dictionary comprehension
print(Comprehensive_dict)
```

```python
even_dictionry = {x : x**2 for x in range(1, 11) if x % 2 == 0} #Creating a dictionary of squares of even numbers from 1 to 10 using dictionary comprehension with a condition
print(even_dictionry)
```

#### Merged dictionary


```python
dictionary_1 = {"name": "Harini", "age": 23, "city": "Vizag", "qualification": "B.Tech"}
dictionary_2 = {"name": "Hari", "age": 24, "city": "Vizag", "qualification": "M.Tech"}
merged_dictionary = {**dictionary_1, **dictionary_2} #Merging two dictionaries using the unpacking operator (**)
print(merged_dictionary)
```

#### Additional functions in list are:

```python
lists = [x for x in range(17)]
print(lists)
```

```python
len_value = len(lists) #Finding the length of the list using the len() function
print(len_value)
```
