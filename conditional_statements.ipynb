### Conditional statements
#### Conditional statements works like real life scenario where if a certain condition is meeting then a statement of code or a block of code gets executed.
#### IF statement
#### ELIF statement
#### ELSE statement

```python
age = int(input("Enter your age: ")) #Taking user input for age and converting it to an integer
if age >= 18: #Checking if the age is greater than or equal to 18
    print("You are eligible to vote.") #If the condition is true, print that the user is eligible to vote
else: #If the condition is false
    print("You are not eligible to vote.") #If the condition is false, print that the user is not eligible to vote
```

```python
#Figure out who is taller among 3 people
height1 = float(input("Enter the height of person 1 : ")) #Taking user input for height of person 1 and converting it to a float
height2 = float(input("Enter the height of person 2 : ")) #Taking user input for height of person 2 and converting it to a float
height3 = float(input("Enter the height of person 3 : ")) #Taking user input for height of person 3 and converting it to a float
if height1 > height2 > height3: #Checking if height of person 1 is greater than both person 2 and person 3
    print("Person 1 is the tallest.") 
elif height2 > height1 > height3: #Checking if height of person 2 is greater than both person 1 and person 3
    print("Person 2 is the tallest.") 
else: #If both previous conditions are false, then person 3 must be the tallest
    print("Person 3 is the tallest.") 
```

### Nested conditional statement
#### It involves IF statements inside another IF statement

```python
#check whetehr the year is a leap year or not
year = int(input("Enter a year: ")) #Taking user input for year and converting it to an integer
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0): #Checking if the year is a leap year using the leap year rules
   print(year, "is a leap year.") #If the condition is true, print that the year is a leap year   
else: #If the condition is false
    print(year, "is not a leap year.") #If the condition is false, print that the year is not a leap year
```
