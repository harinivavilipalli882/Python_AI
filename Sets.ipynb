```python
set_fruits = {"apple", "banana", "cherry", "grape", "kiwi"}
fruits_vegetables = {"apple", "banana", "carrot", "broccoli", "spinach", "kale", "lettuce"}

for fruit in set_fruits:
    fruits_vegetables.add(fruit)
print(fruits_vegetables)
```
