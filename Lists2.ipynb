```python
lists = ["mango","apple","Guava","orange","grapes","banana"]
print(lists[5])
```

```python
print(lists[3])
```

#### Negative indexing
##### In any data structure we can access the data using negative indexing as well

```python
lists[-1]
```

```python
lists[-5]
```

#### List slicing
##### For having access for all of the values we are going to store inside the list, we can use :
##### List slicing can help us out with retrieving specific data points from our list.

```python
print(lists[:])
```

```python
lists[1:4]
```

```python
lists[1:3]
```

```python
lists[::-1]       #Reverse a list using negative indexing
```

```python
lists[0:]        #to print all value from starting
```

#### Lists are Mutable
##### Mutable means we can replace the values present at a specific location inside our list

```python
vegetables = ["potato","tomato","onion","beans","carrot"]
```

```python
vegetables[2]
```

```python
vegetables[2] = "garlic"
```

```python
print(vegetables)
```

#### By using looping statements, we can iterate through list

```python
for veggie in vegetables :
    print(veggie)
```

```python
list_item = [8,25,"Hari",23.0,True,"Harini"]
for i in list_item:
    print(i)
```

```python
list_item[::2]     #gives the values from starting with a space of two
```

```python
list_item[::-2]   #gives the values from last value, with a space of two
```
